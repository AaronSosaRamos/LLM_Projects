```python
from pydantic import BaseModel, Field
from typing import List, Dict
from datetime import date

class AssetAllocation(BaseModel):
    asset_class: str = Field(..., description="Asset class (e.g., 'Equity', 'Fixed Income', 'Commodity')")
    percentage: float = Field(..., description="Percentage of the portfolio allocated to this asset class")

class PortfolioRiskMetrics(BaseModel):
    volatility: float = Field(..., description="Annualized portfolio volatility")
    beta: float = Field(..., description="Portfolio beta relative to a benchmark index")
    value_at_risk: float = Field(..., description="Value at Risk (VaR) at a given confidence level")
    sharpe_ratio: float = Field(..., description="Portfolio Sharpe Ratio")
    sortino_ratio: float = Field(..., description="Portfolio Sortino Ratio")
    maximum_drawdown: float = Field(..., description="Maximum observed loss from a peak to a trough")

class TechnicalIndicator(BaseModel):
    name: str = Field(..., description="Name of the technical indicator (e.g., 'SMA', 'RSI')")
    values: Dict[str, float] = Field(..., description="Date and corresponding value of the indicator for the given period")
    signal: str = Field(..., description="Indicator signal based on its value (e.g., 'buy', 'sell', 'hold')")
    trend: str = Field(..., description="Indicates whether the stock is trending upward or downward based on the indicator")

class FinancialMetric(BaseModel):
    metric_name: str = Field(..., description="Name of the financial metric (e.g., 'PE Ratio', 'EPS')")
    metric_value: float = Field(..., description="The value of the financial metric")
    industry_average: float = Field(..., description="Industry average for comparison")
    analysis: str = Field(..., description="Analysis of the metric value (e.g., 'overvalued', 'undervalued')")
    percentile_rank: float = Field(..., description="Percentile rank of the stock based on the metric compared to its peers")

class SentimentAnalysis(BaseModel):
    source: str = Field(..., description="Source of the news or social media sentiment (e.g., 'Reuters', 'Twitter')")
    positive_mentions: int = Field(..., description="Number of positive mentions of the stock in the news")
    negative_mentions: int = Field(..., description="Number of negative mentions of the stock in the news")
    neutral_mentions: int = Field(..., description="Number of neutral mentions of the stock in the news")
    sentiment_score: float = Field(..., description="Overall sentiment score for the stock from -1 (negative) to 1 (positive)")
    sentiment_trend: str = Field(..., description="Indicates whether sentiment is improving or worsening")

class RiskAssessment(BaseModel):
    volatility: float = Field(..., description="Stock price volatility during the analysis period (measured by standard deviation)")
    beta: float = Field(..., description="Stock's beta value compared to the market index")
    risk_level: str = Field(..., description="Risk level of the stock (e.g., 'low', 'medium', 'high')")
    recommended_exposure: str = Field(..., description="Recommended exposure level to the stock based on the risk profile (e.g., 'reduce', 'hold', 'increase')")
    tail_risk: float = Field(..., description="Potential loss in extreme downside scenarios (e.g., during market crashes)")

class PerformanceOverview(BaseModel):
    start_price: float = Field(..., description="Stock price at the start of the analysis period")
    end_price: float = Field(..., description="Stock price at the end of the analysis period")
    price_change_percentage: float = Field(..., description="Percentage change in stock price during the analysis period")
    dividend_yield: float = Field(..., description="Dividend yield during the analysis period")
    total_return_percentage: float = Field(..., description="Total return (including dividends) during the analysis period")
    market_cap: float = Field(..., description="Market capitalization at the end of the analysis period")
    pe_ratio: float = Field(..., description="Price-to-Earnings ratio at the end of the analysis period")
    earnings_growth_rate: float = Field(..., description="Rate of earnings growth during the analysis period")

class StockAnalysis(BaseModel):
    stock_symbol: str = Field(..., description="Stock symbol of the company")
    company_name: str = Field(..., description="Full name of the company")
    sector: str = Field(..., description="Sector in which the company operates")
    industry: str = Field(..., description="Industry in which the company operates")
    performance_overview: PerformanceOverview = Field(..., description="Overview of the stock's performance")
    technical_indicators: List[TechnicalIndicator] = Field(..., description="List of technical indicators")
    financial_metrics: List[FinancialMetric] = Field(..., description="List of financial metrics")
    sentiment_analysis: List[SentimentAnalysis] = Field(..., description="List of sentiment analysis results")
    risk_assessment: RiskAssessment = Field(..., description="Risk assessment of the stock")
    recommendation: str = Field(..., description="Recommendation for the stock (e.g., 'buy', 'hold', 'sell')")
    allocation_suggestion: float = Field(..., description="Suggested allocation percentage for this stock in the portfolio")
    notes: str = Field(..., description="Additional notes or comments on the stock")

class PortfolioAnalysisOutput(BaseModel):
    portfolio_name: str = Field(..., description="Name of the investment portfolio")
    analysis_date: date = Field(..., description="Date when the analysis was performed")
    total_value: float = Field(..., description="Total value of the portfolio")
    asset_allocation: List[AssetAllocation] = Field(..., description="Current asset allocation of the portfolio")
    risk_metrics: PortfolioRiskMetrics = Field(..., description="Risk metrics for the overall portfolio")
    stocks_analysis: List[StockAnalysis] = Field(..., description="Detailed analysis for each stock in the portfolio")
    overall_recommendation: str = Field(..., description="Overall investment recommendation for the portfolio")
    suggested_rebalancing_actions: List[str] = Field(..., description="List of suggested actions to rebalance the portfolio")
    expected_return: float = Field(..., description="Expected return of the portfolio based on the analysis")
    confidence_level: str = Field(..., description="Confidence level of the analysis (e.g., 'high', 'medium', 'low')")
    notes: str = Field(..., description="Additional notes or comments on the portfolio")
```

```python
from crewai_tools import tool

class CalculatorTools():

  @tool("Make a calculation")
  def calculate(operation: str) -> str:
    """Useful to perform any mathematical calculations,
    like sum, minus, multiplication, division, etc.
    The input to this tool should be a mathematical
    expression, a couple examples are `200*7` or `5000/2*10`
    """
    return eval(operation)
```

```python
from crewai import Agent
from langchain_community.tools import TavilySearchResults
from langchain.tools import Tool
from langchain_openai import ChatOpenAI
from langchain_community.utilities import GoogleSerperAPIWrapper

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0).bind(response_format={"type": "json_object"})
financial_search = GoogleSerperAPIWrapper()

compiled_tools = [
    CalculatorTools.calculate,
    TavilySearchResults(
        max_results=5,
        include_answer=True,
        include_raw_content=True
    ),
    Tool(
        name="Financial News Extractor",
        func=financial_search.run,
        description="Useful for fetching the latest financial news and reports for a given company or sector.",
    ),
    Tool(
        name="Stock Price Tracker",
        func=financial_search.run,
        description="Fetches the current and historical stock prices for a given stock symbol.",
    ),
    Tool(
        name="Market Sentiment Analyzer",
        func=financial_search.run,
        description="Analyzes the market sentiment for a company or sector based on recent news articles.",
    )
]

class PortfolioAnalysisAgents:
    def market_analyst(self):
        return Agent(
            role='Market Analyst',
            goal="""Provide in-depth market data and trend analysis to inform portfolio decisions""",
            backstory="""An experienced market analyst with expertise in macroeconomic trends and their impact on investment portfolios.""",
            verbose=True,
            tools=compiled_tools,
            llm=llm,
            memory=True
        )

    def financial_analyst(self):
        return Agent(
            role='Financial Analyst',
            goal="""Analyze the financial health and performance of each stock in the portfolio""",
            backstory="""A seasoned financial analyst skilled in evaluating company financials and identifying investment opportunities.""",
            verbose=True,
            tools=compiled_tools,
            llm=llm,
            memory=True
        )

    def portfolio_strategist(self):
        return Agent(
            role='Portfolio Strategist',
            goal="""Synthesize analyses to provide strategic portfolio recommendations""",
            backstory="""An expert in portfolio management, optimizing asset allocation to meet investment objectives and risk tolerance.""",
            verbose=True,
            tools=compiled_tools,
            llm=llm,
            memory=True
        )
```

```python
from crewai import Task
from textwrap import dedent
from datetime import date

class PortfolioAnalysisTasks:
    def gather_market_data(self, agent, input_data):
        return Task(
            description=dedent(f"""
                Collect and summarize recent market data, economic indicators, and trends that could impact the portfolio. Focus on factors like interest rates, inflation, GDP growth, and geopolitical events.

                Pay attention to how these macroeconomic factors affect the sectors and industries represented in the portfolio.

                Ensure that the data is up-to-date and relevant to the analysis period from {input_data['analysis_period_start']} to {input_data['analysis_period_end']}.
                """),
            expected_output="""
                Provide a comprehensive report on current market conditions, trends, and potential impacts on the portfolio. Include any notable events or shifts that could influence investment decisions.
                """,
            agent=agent
        )

    def analyze_stocks_financials(self, agent, input_data):
        return Task(
            description=dedent(f"""
                For each stock in the portfolio ({', '.join(input_data['stocks'])}), perform a thorough financial analysis.

                Examine key financial metrics such as P/E ratio, EPS growth, revenue trends, debt-to-equity ratio, and other relevant indicators.

                Compare each stock's performance to its industry peers and overall market trends.

                Consider the company's financial health, strengths, weaknesses, and outlook.

                Use the most recent financial statements and data available.
                """),
            expected_output="""
                Provide a detailed financial analysis for each stock, including assessments of their financial standing, competitive position, and potential risks and opportunities.
                """,
            agent=agent
        )

    def assess_portfolio_risk(self, agent, input_data):
        return Task(
            description=dedent(f"""
                Evaluate the overall risk of the portfolio.

                Calculate risk metrics such as portfolio volatility, beta, Value at Risk (VaR), Sharpe ratio, Sortino ratio, and maximum drawdown.

                Analyze the diversification of the portfolio across asset classes, sectors, and industries.

                Identify any concentrations or correlations that may increase risk.

                Consider the investor's risk tolerance ({input_data['risk_tolerance']}) and investment objectives ({input_data['investment_goals']}).
                """),
            expected_output="""
                Provide a comprehensive risk assessment of the portfolio, highlighting potential risks, their implications, and suggestions for mitigation.
                """,
            agent=agent
        )

    def develop_portfolio_strategy(self, agent, input_data):
        portfolio_analysis_output_schema = PortfolioAnalysisOutput.schema_json(indent=2)

        return Task(
            description=dedent(f"""
                Review and synthesize the analyses provided by the Market Analyst and Financial Analyst.

                Develop strategic recommendations for the portfolio.

                This should include suggestions for asset allocation adjustments, stock additions or removals, and rebalancing actions.

                Consider the investor's profile ({input_data['investor_profile']}), risk tolerance ({input_data['risk_tolerance']}), and investment goals ({input_data['investment_goals']}).

                Provide an overall recommendation and expected return for the portfolio.

                Your response must be a valid JSON matching the following schema:
                {portfolio_analysis_output_schema}

                Ensure that the JSON is properly formatted and complete.
                """),
            expected_output=f"""
                Provide a detailed portfolio strategy report in JSON format according to the provided schema: {portfolio_analysis_output_schema}, including all required fields and recommendations.
                Ensure that the JSON is properly formatted and complete.
                """,
            agent=agent
        )
```

```python
from crewai import Crew
from textwrap import dedent
from datetime import datetime

class PortfolioAnalysisCrew:
    def __init__(self, input_data):
        self.input_data = input_data

    def run(self) -> str:
        agents = PortfolioAnalysisAgents()
        tasks = PortfolioAnalysisTasks()

        market_analyst_agent = agents.market_analyst()
        financial_analyst_agent = agents.financial_analyst()
        portfolio_strategist_agent = agents.portfolio_strategist()

        market_data_task = tasks.gather_market_data(market_analyst_agent, self.input_data)
        financial_analysis_task = tasks.analyze_stocks_financials(financial_analyst_agent, self.input_data)
        risk_assessment_task = tasks.assess_portfolio_risk(portfolio_strategist_agent, self.input_data)
        strategy_task = tasks.develop_portfolio_strategy(portfolio_strategist_agent, self.input_data)

        crew = Crew(
            agents=[
                market_analyst_agent,
                financial_analyst_agent,
                portfolio_strategist_agent
            ],
            tasks=[
                market_data_task,
                financial_analysis_task,
                risk_assessment_task,
                strategy_task
            ],
            verbose=True
        )

        result = crew.kickoff()
        return result

if __name__ == "__main__":
    print("## Welcome to the Portfolio Investment Analysis Tool")
    print('-------------------------------')

    portfolio_name = input("Enter the name of the investment portfolio: ")
    stocks_input = input("Enter the stock symbols to analyze, separated by commas: ")
    analysis_period_start = input("Enter the analysis period start date (YYYY-MM-DD): ")
    analysis_period_end = input("Enter the analysis period end date (YYYY-MM-DD): ")
    investor_profile = input("Enter the investor profile (e.g., 'conservative', 'aggressive'): ")
    risk_tolerance = input("Enter the investor's risk tolerance level (e.g., 'low', 'medium', 'high'): ")
    investment_goals = input("Enter the investor's investment goals (e.g., 'long-term growth', 'diversification'): ")

    # Convert dates
    analysis_period_start = datetime.strptime(analysis_period_start, '%Y-%m-%d').date()
    analysis_period_end = datetime.strptime(analysis_period_end, '%Y-%m-%d').date()

    # Process stock list
    stocks = [stock.strip().upper() for stock in stocks_input.split(',')]

    input_data = {
        'portfolio_name': portfolio_name,
        'stocks': stocks,
        'analysis_period_start': analysis_period_start,
        'analysis_period_end': analysis_period_end,
        'investor_profile': investor_profile,
        'risk_tolerance': risk_tolerance,
        'investment_goals': investment_goals
    }

    portfolio_crew = PortfolioAnalysisCrew(input_data)

    result = portfolio_crew.run()
    print("\n\n########################")
    print("## Here is the Report")
    print("########################\n")
    print(result)
```
