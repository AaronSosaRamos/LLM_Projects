```python
import os
from google.colab import userdata

os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
```

```python
from openai import OpenAI
client = OpenAI()

response = client.chat.completions.create(
  model="o1-mini",
  messages=[
        {
            "role": "user",
            "content": "Can you show me how to develop a multithreading architecture with microservices using Spring?"
        }
  ]
)
```

```python
response
```

```python
print(response.choices[0].message)
```
